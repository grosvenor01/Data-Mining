```python
to_numeric = ["Acc_x","Acc_y","Acc_z","Gyro_x","Gyro_y","Gyro_z","Set"]
for i in to_numeric:
    df[i]=pd.to_numeric(df[i],errors="coerce")
df= df.dropna()
```

```python
def hamming(string1 , string2):
    minimum = min(len(string1) , len(string2))
    count = 0 
    for i in range(minimum):
        if string1[i] != string2[i]:
            count+=1
    return count + abs(len(string1) - len(string2))

def manhatten(atr1 , atr2):
    return abs(atr1 - atr2)
```

```python
columns = df.columns[:-1]
X = ["2024-11-20 18:09:51.000", -0.137, 1.066, 0.8215, -6.597, 0.808, 1.985, "B", "medium" ,"", 30]
def distance(df,index):
    dist = 0
    for i, col in enumerate(columns):

        if df[col].dtype == "object":
            try:
                dist += hamming(df.iloc[index][col], X[i])
            except ValueError:
                print(f"Warning: Hamming distance error for column '{col}'. Treating as max distance.")
        else:
            try:
                dist += manhatten(df.iloc[index][col], X[i])
            except ValueError:
                print(f"Warning: Hamming distance error for column '{col}'. Treating as max distance.")
    return dist

distance(df , 0)
```

```python
distances = []
for i in range(len(df)):
    dist = distance(df, i)
    distances.append(dist)
df['distance'] = distances
```

```python
df = df.sort_values(by="distance")
df
```
