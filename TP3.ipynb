```python
to_numeric = ["Acc_x","Acc_y","Acc_z","Gyro_x","Gyro_y","Gyro_z","Set"]
for i in to_numeric:
    df[i]=pd.to_numeric(df[i],errors="coerce")
df= df.dropna()
df.head()
```

1. descretizing Data & Replacing with the average

```python
def labeling(df , intervals , attribute):
    avg = []
    for i in range(len(intervals)-1) :
        df_cals = df[(df[attribute]>intervals[i]) & (df[attribute]<intervals[i+1])]
        avg.append(df_cals[attribute].mean())
    return avg

def descrtizing(df , attribute):
    try : 
        intervals = []
        k = int(1 + (10/3) * math.log10(len(df)))
        max_val = max(df[attribute])
        min_val = min(df[attribute])

        width = (max_val - min_val) / k
        for i in range(k+1):
            intervals.append(min_val)
            min_val+=width
        print("============ INTERVALS ============")
        print(intervals)
        # Store labels as average values ? 
        avg = labeling(df , intervals , attribute)
        
        print("============ AVERAGE ============")
        print(avg)
        df[f"Category {attribute}"]  = pd.cut(df[attribute] , intervals , labels = avg)
        print("============ INFOs ============")
        print(df[f"Category {attribute}"].value_counts())
        print("============ DATA FRAME ============")
        return df
    except Exception as e: 
        print("non numerical values detected")
        print(e)

descrtizing(df , attribute)
```

3. Normalization

```python
def normalization(x , minimum , maximum):
    normalized = (x - minimum)/(maximum - minimum)
    return normalized
columns  = [i for i in df.columns if df[i].dtype == "float64"]
for i in columns : 
    df[i] = df[i].apply(lambda x : normalization(x , min(df[i]) , max(df[i])))
df
```
